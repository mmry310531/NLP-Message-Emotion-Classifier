```python
from datasets import load_dataset
dataset = load_dataset('emotion', split='train')
```

```python
df = dataset.to_pandas()
```

```python
df.head()
# tweets dataset      final
# 0 sad           ->  0 sad
# 1 joy           ->. 6 joy
# 2 love          ->  #
# 3 anger         ->  22 angry
# 4 fear          ->  11 afraid
# 5 surprise      ->  9 surprised
```

```python
df = df[df['label'] != 2]
```

```python
mapping = {
    0 : 0,
    1 : 6, 
    3 : 22, 
    4 : 11, 
    5 : 9
}
df['label'] = df['label'].map(mapping)
df.to_csv("tweetData.csv")
```

```python
data_train = pd.read_csv("new_train.csv")
data_train['text'] = data_train['prompt'] + " " + data_train['utterance']
data_train = data_train[['text', 'label']]
```

```python
result = pd.concat([data_train, df], ignore_index=True)
result.to_csv("NewTrainData.csv")
result
```
